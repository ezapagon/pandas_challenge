```python

import pandas as pd
import os
```

```python
# Save path to data set in a variable
data_file_schools_path = "/Users/Edo/Resources/schools_complete.csv"
data_file_schools_path
```

```python
data_file_students_path = "/Users/Edo/Resources/students_complete.csv"
data_file_students_path
```

```python
# Use Pandas to read data
data_file_schools = pd.read_csv(data_file_schools_path)
data_file_schools.head()
```

```python
# Use Pandas to read data
data_file_students = pd.read_csv(data_file_students_path)
data_file_students.head()
```

```python
total_schools = data_file_schools["School ID"].count()
total_students = data_file_students["Student ID"].count()
total_budget = data_file_schools["budget"].sum()
average_math_score = data_file_students["math_score"].mean()
average_reading_score = data_file_students["reading_score"].mean()
passing_math = data_file_students.query('math_score >=70')["math_score"].count()/total_students*100
passing_reading = data_file_students.query('reading_score >=70')["reading_score"].count()/total_students*100
overall_passing = (average_math_score + average_reading_score)/2

painting_school_df = pd.DataFrame({
      "Total Schools":[total_schools],
      "Total Students":[total_students],
      "Total Budget":[total_budget],
      "Average Math Score":[average_math_score],
      "Average Reading Score":[average_reading_score],
      "% Passing Math":[passing_math],
      "% Passing Reading":[passing_reading],
      "% Overall Passing":[overall_passing]
})
painting_school_df
```

```python
school_data_complete = pd.merge( data_file_schools, data_file_students, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
overview_school_data = school_data_complete.groupby(['school_name',"type","budget"])
overview_school_data

overview_total_schools = overview_school_data["School ID"].count()
overview_average_math_score = overview_school_data["math_score"].mean()
overview_average_reading_score = overview_school_data["reading_score"].mean()


data = {"Total Students":overview_total_schools,
        "Average Math Score":overview_average_math_score,
         "Average Reading Score":overview_average_reading_score,
         
}
df = pd.concat(data,axis=1)

df



```
